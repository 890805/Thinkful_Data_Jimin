```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("data_cleaning.csv")
```

```python
top_5_journal = df.groupby('Journal title').count().sort_values('Article title', ascending = False).head(5)
```

```python
top_5_journal = top_5_journal.reset_index()
top_5_journal
```

```python
top_5_journal = list(top_5_journal['Journal title']) # the five most common journals 
```

```python
top_5_df = df[df['Journal title'].isin(top_5_journal)] # the total articles for the five most common journals 
top_5_df
```

```python
top_5_df['COST (£) charged to Wellcome (inc VAT when charged)'] = top_5_df['COST (£) charged to Wellcome (inc VAT when charged)'].str[1:].astype('float')
```

```python
top_5_df.groupby('Journal title').mean()
```

```python
top_5_df.groupby('Journal title').median()
```

```python
top_5_df.groupby('Journal title').std()
```

```python
sns.distplot(top_5_df['COST (£) charged to Wellcome (inc VAT when charged)'])
plt.show()
```

### Some values seem to be too extream based on mean and std.

### According to Internet Search, generally, the submission fee is not over 10000.

```python
new_df = top_5_df[(top_5_df['COST (£) charged to Wellcome (inc VAT when charged)']  < 10000)]
```

```python
sns.distplot(new_df['COST (£) charged to Wellcome (inc VAT when charged)'])
plt.show()
```

```python
new_df.groupby('Journal title').mean()
```

```python
new_df.groupby('Journal title').median()
```

```python
new_df.groupby('Journal title').std()
```
