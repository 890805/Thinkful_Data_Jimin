```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
df
```

## HW1.1 Choose one variable and plot that variable four different ways.

### 1) Histogram
* Pros: Viewrs can easily compare data. It's useful when dealing with large value ranges.

* Cons: The results can be different by interval setting.

```python
## Histogram
sns.set(style="white", color_codes=True)
f, axes = plt.subplots(2, 2, figsize=(10, 10), sharex=True)
sns.distplot(df['radius_mean'], kde=False, color="b", ax=axes[0, 0])
sns.distplot(df['radius_mean'], hist=False, rug=True, color="r", ax=axes[0, 1])
sns.distplot(df['radius_mean'], hist=True, color="g", ax=axes[1, 0])
sns.distplot(df['radius_mean'], hist=False, color="m", kde_kws={"shade": True}, ax=axes[1, 1])
sns.despine()
plt.tight_layout()
plt.show()
```

### 2) Box plot

* Pros: It is easy to read minimum, median, outliers, quantiles, and maximum value. We can find skweness.
* Cons: We cannot identify mean and mode value, so it can be hard to figure out the exact shape of the data.

```python
## Box plot
sns.boxplot( y='radius_mean', data=df,)
sns.despine()
plt.show()
```

### 3) Violin plot

It is similar with Box plot

```python
## Violin plot
sns.violinplot(y='radius_mean', data=df, estimator=np.std)
sns.despine()
plt.show()
```

### 4) Point plot

This type of plot is better to use at comparing two variables.

```python
## Point Plot
sns.pointplot(x="radius_mean", data=df, orient='v')
sns.despine()
plt.show()
```

### 5) Swarm plot

* Pros: We can recognize individual data points and disribution shape.
* Cons: It is hard to deal with the large size of data.

```python
## Swarm plot
sns.swarmplot(y='radius_mean', data=df)
sns.despine()
plt.show()
```

## HW1.2 Choose two continuous variables, and plot them three different ways.

### 1) Hexbin plot

```python
## Hexbin plot
sns.jointplot(x='radius_mean',y='perimeter_mean',data=df,kind='hex')
sns.despine()
plt.show()
```

### 2) Lm plot

```python
##lm plot
sns.jointplot(x='radius_mean',y='perimeter_mean', data=df,kind='reg')
sns.despine()
plt.show()
```

### 3) Kde Plot

```python
## Kde plot
sns.jointplot(x='radius_mean',y='perimeter_mean',data=df,kind='kde')
sns.despine()
plt.show()
```

## HW1.3 Choose one continuous variable and one categorical variable, and plot them six different ways.

### 1) Bar plot

```python
sns.barplot(x='diagnosis', y='radius_mean', data = df, estimator=np.std, palette="RdBu_r")
sns.despine()
plt.show()
```

### 2) Box plot

```python
sns.boxplot(x='diagnosis', y='radius_mean', data = df, palette="RdBu_r")
sns.despine()
```

### 3) Violin plot

```python
sns.violinplot(x='diagnosis', y='radius_mean', data=df, palette="RdBu_r")
sns.despine()
plt.show()
```

### 4) Strip plot

```python
sns.stripplot(x='diagnosis', y='radius_mean', data=df, palette="RdBu_r", jitter=False)
sns.despine()
plt.show()
```

### 5) Swarm plot

```python
sns.swarmplot(x='diagnosis', y='radius_mean', data=df, palette="RdBu_r")
sns.despine()
plt.show()
```

### Combined plot Violin + Swarm plot

```python
sns.violinplot(x='diagnosis', y='radius_mean', data=df, palette="RdBu_r", jitter=False)
sns.swarmplot(x='diagnosis', y='radius_mean', color = 'gray', size = 3, data=df)
sns.despine()
plt.show()
```

### 7) Boxen plot

```python
sns.boxenplot(x='diagnosis', y='radius_mean', data=df)
sns.despine()
plt.show()
```
