## Assignments

For each of the following questions, outline how you could use an A/B test to find an answer. Be sure to identify all five key components of an A/B test we outlined above.

**Five key components of an A/B test**

1. Two versions of something whose effects will be compared. 

2. A sample, divided into two groups.

3. A hypothesis.

4. Outcome(s) of interest.

5. Other measured variables. 


###  1

**Does a new supplement help people sleep better?**

1. Two versions of something whose effects will be compared.

    1) Control supplements without new supplement	
    2) Treatment supplements with new supplement
    
2. Sample: recruit patients who were diagnosed with a sleep disorder from one hospital.

3. A hypothesis: new supplement will help the patients sleep longer.

4. Outcome(s) of interest: change of sleep hours of patients

5. Other measured variables: age, sex, or diet habit, etc.

### 2

**Will new uniforms help a gym's business?**

1. Two versions of something whose effects will be compared.

    1) Control: employees who wear old uniforms
	
    2) Treatment: employees who wear new uniforms
    
2. Sample: the records of new registration for two month. (1 month for control, 1 month for treat)

3. A hypothesis: new uniforms will help a gym's new registration rate.

4. Outcome(s) of interest: the number of new registrations.

5. Other measured variables: the list of employee

4. Outcome(s) of interest:

5. Other measured variables: visiting time

### 4

**If I put 'please read' in the email subject will more people read my emails?**

1. Two versions of something whose effects will be compared.

    1) Control: email subjects including 'please read' message.
	
    2) Treatment: email subjects not including 'please read' message.
     
2. Sample: randomly send one of the two versions of the email to a subset of subscribers to our newsletter 

3. A hypothesis: the group who get the email that includes 'please read' in the subject will more open the email than another group.

4. Outcome(s) of interest: whether people open emails or not

5. Other measured variables: age, gender, etc.
